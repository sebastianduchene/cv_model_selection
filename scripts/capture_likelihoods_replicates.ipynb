```r
get_likelihood <- function(output){
    as.numeric(gsub(' |[A-Z]|[a-z]|[(]|[)][.]|[a-z][.][a-z]', '', grep('Tree.+', output, value = T)))
}
```

```r
test_sets <- dir('enterovirus/', pattern = 'test.+fasta')
```

```r
#dir('enterovirus/', pattern = 'log|trees')
```

```r
trees_sc <- gsub('.fasta', '_strict_constant_phylogs.trees', gsub('test', 'training', test_sets))
logs_sc <- gsub('.fasta', '_strict_constant.log', gsub('test', 'training', test_sets))

trees_se <- gsub('.fasta', '_strict_exponential_phylogs.trees', gsub('test', 'training', test_sets))
logs_se <- gsub('.fasta', '_strict_exponential.log', gsub('test', 'training', test_sets))

# 
trees_uc <- gsub('.fasta', '_ucld_constant_phylogs.trees', gsub('test', 'training', test_sets))
logs_uc <- gsub('.fasta', '_ucld_constant.log', gsub('test', 'training', test_sets))

trees_ue <- gsub('.fasta', '_ucld_exponential_phylogs.trees', gsub('test', 'training', test_sets))
logs_ue <- gsub('.fasta', '_ucld_exponential.log', gsub('test', 'training', test_sets))
```

```r
setwd("enterovirus/")
# THis needs to be done for each of the four models. Or in the four loops, evaluate all the models for each test set

out_matrix <- matrix(NA, 10, 4)
colnames(out_matrix) <- c('sc', 'se', 'uc', 'ue')

out_sc <- vector()
for(i in 1:length(logs_sc)){
    print(c(logs_sc[i], trees_sc[i]))
    command <- paste('python ../cv_p4_likelihoods.py', test_sets[i], logs_sc[i], trees_sc[i])
    out_matrix[i, 1] <- mean(get_likelihood(system(command, intern = T)))

    command <- paste('python ../cv_p4_likelihoods.py', test_sets[i], logs_se[i], trees_se[i])
    out_matrix[i, 2] <- mean(get_likelihood(system(command, intern = T)))

    command <- paste('python ../cv_p4_likelihoods.py', test_sets[i], logs_uc[i], trees_uc[i])
    out_matrix[i, 3] <- mean(get_likelihood(system(command, intern = T)))

    command <- paste('python ../cv_p4_likelihoods.py', test_sets[i], logs_ue[i], trees_ue[i])
    out_matrix[i, 4] <- mean(get_likelihood(system(command, intern = T)))


    
}

print(out_matrix)

setwd('..')
```

```r
print(colMeans(out_matrix))
```
