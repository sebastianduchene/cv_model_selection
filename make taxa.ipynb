```r
library(TreeSim)
library(phangorn)
```

```r
tr <- rtree(50)
tr$tip.label <- paste0('t', 1:50, '_', c(0, sample(1:50, replace = T, size = 49)))
```

```r
plot(tr)
```

```r
s <- simSeq(tr, l = 1000)
```

```r
write.dna(as.DNAbin(s), file =  'empty_dat.fasta', format = 'fasta', nbcol = -1, colsep = '')
```
