```r
library(ape)
```

```r
all_files <- dir(pattern = '_ACL_.+nex$')
original_files <- all_files[-grep('new|train', all_files)]
```

```r
for(f in original_files){
    new_name <- paste0(gsub('nex|_ACL', '', f), 'fasta')
    temp_data <- as.DNAbin(read.nexus.data(f))
    write.dna(temp_data, new_name, format = 'fasta', nbcol = -1, colsep = '')
    print(new_name)
}
```
