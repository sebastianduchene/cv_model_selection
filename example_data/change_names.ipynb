```r
library(ape)
```

```r
s4_data <- dir(pattern = 's4_(strict|exp|ucld)[.]fasta')
```

```r
s4_data
```

```r
for(f in s4_data){
    dat <- read.dna(f, format = 'fasta')
    training <- dat[, 1:(ncol(dat)/2)]
    test <- dat[, (ncol(dat)/2):ncol(dat)]
    
    write.dna(training, file = gsub('[.]fasta', '_training.fasta', f), format = 'fasta',
             nbcol = -1, colsep = '')

    write.dna(test, file = gsub('[.]fasta', '_test.fasta', f), format = 'fasta',
             nbcol = -1, colsep = '')
}
```
