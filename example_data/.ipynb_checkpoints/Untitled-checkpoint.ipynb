```r
parse_p4_ouput <- function(x){
    liks <- grep('^Tree.calcLog', x, value = T)
    out <- as.numeric(gsub('[A-Z]|[a-z]|[a-z][.][a-z]|[(]|[)]|[.] ', '', liks))
    return(out)
}
```

```r
x <- system('python ../scripts/cv_p4_likelihoods_JC.py s4_ucld_test.fasta s4_ucld_subset_strict_phylogs.trees', intern = T)
sc_liks <- parse_p4_ouput(x)
```

```r
x <- system('python ../scripts/cv_p4_likelihoods_JC.py s4_ucld_test.fasta s4_ucld_subset_ucld_phylogs.trees', intern = T)
ucld_liks <- parse_p4_ouput(x)
```
