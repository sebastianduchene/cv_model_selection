```r
library(ape)
library(TreeSim)
library(phangorn)
library(NELSI)
```

```r
trees <- list()

for(i in 1:100){
    tr <- sim.bd.taxa.age(n = 50, lambda = 1, mu = 0, age = 100, numbsim = 1)[[1]]
    tr$edge.length <- (tr$edge.length / max(allnode.times(tr))) * 100
    
    #strict
    tr_s <- tr
    tr_s$edge.length <- tr$edge.length * 0.001
    ss <- simSeq(tr_s)
    write.dna(as.DNAbin(ss), file= paste0('s', i, '_strict.fasta'), format = 'fasta', nbcol = -1, colsep = '')
    write.dna(as.DNAbin(ss)[, 1:500], file= paste0('s', i, '_strict_subset.fasta'), format = 'fasta', nbcol = -1, colsep = '')
    
    #ucl
    tr_ucl <- tr
    tr_ucl$edge.length <- tr$edge.length * rlnorm(length(tr_ucl$edge.length), meanlog = log(0.001), sd = 1)
    sucld <- simSeq(tr_ucl)
    write.dna(as.DNAbin(sucld), file = paste0('s', i, '_ucld.fasta'), format = 'fasta', nbcol = -1, colsep= '')
    write.dna(as.DNAbin(sucld)[, 1:500], file = paste0('s', i, '_ucld_subset.fasta'), format = 'fasta', nbcol = -1, colsep = '')
    
    #exp
    tr_exp <- tr
    tr_exp$edge.length <- tr$edge.length * rexp(length(tr_exp$edge.length), 1/0.001)
    sexp <- simSeq(tr_exp)
    write.dna(as.DNAbin(sexp), file = paste0('s', i, '_exp.fasta'), format = 'fasta', nbcol = -1, colsep = '')
    write.dna(as.DNAbin(sexp)[, 1:500], file = paste0('s', i, '_exp_subset.fasta'), format = 'fasta', nbcol = -1, colsep = '')
}
```
