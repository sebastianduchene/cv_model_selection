```r
# Make training indices for these data
dir()
acl_1 <- read.dna('ACL_1.fasta', format = 'fasta')
args(make_mrbayes_block)
```

```r
set.seed(1234)
training <- sample(1:ncol(acl_1),size = 1000)
save(training, file = 'training_indicies_sims2.Rdata')
```

```r
fasta_files <- dir(pattern = 'fasta')

# Need to make 6 files: 3 clocks ss, and trainig. 
for(f in fasta_files){
    complete_data <- read.dna(f, format = 'fasta')
    training_data <- complete_data[, training]
    print(f)
    strict_ss_name <- gsub('.fasta', '_SC_ss.nexus', f)
    make_mrbayes_block(seq_data = complete_data, file_name = strict_ss_name, clock_model = 'strict', ss = T)
    
    igr_ss_name <- gsub('.fasta', '_IGR_ss.nexus', f)
    make_mrbayes_block(seq_data = complete_data, file_name = igr_ss_name, clock_model = 'igr', ss = T)
    
    acl_ss_name <- gsub('.fasta', '_ACL_ss.nexus', f)
    make_mrbayes_block(seq_data = complete_data, file_name = acl_ss_name, clock_model = 'tk02', ss = T)
    
    strict_train_name <- gsub('.fasta', '_SC_train.nexus', f)
    make_mrbayes_block(seq_data = training_data, file_name = strict_train_name, clock_model = 'strict', ss = F)
    
    igr_train_name <- gsub('.fasta', '_IGR_train.nexus', f)
    make_mrbayes_block(seq_data = training_data, file_name = igr_train_name, clock_model = 'igr', ss = F)
    
    acl_train_name <- gsub('.fasta', '_ACL_train.nexus', f)
    make_mrbayes_block(seq_data = training_data, file_name = acl_train_name, clock_model = 'tk02', ss = F)
}
```
