```r
library(ape)
library(TreeSim)
library(phangorn)
library(NELSI)
```

```r
tr1 <- sim.bd.taxa.age(n = 10, lambda = 1, mu = 0, age = 100, numbsim = 1)[[1]]
tr1$edge.length <- (tr1$edge.length / max(allnode.times(tr1))) * 100
tr2 <- sim.bd.taxa.age(n = 10, lambda = 1, mu = 0, age = 100, numbsim = 1)[[1]]
tr2$edge.length <- (tr2$edge.length / max(allnode.times(tr2))) * 100
```

```r
par(mfrow = c(2, 1))
plot(tr1)
nodelabels(round(intnode.times(tr1), 2))

plot(tr2)
nodelabels(round(intnode.times(tr2)))
```

```r
#strict
p1 <- tr1
p1$edge.length <- tr1$edge.length * 0.001
s1 <- as.DNAbin(simSeq(p1))
write.dna(s1, file = 's1_strict.fasta', format = 'fasta', nbcol = -1, colsep = '')

p2 <- tr2
p2$edge.length <- tr2$edge.length * 0.001
s2 <- as.DNAbin(simSeq(p2))
write.dna(s2, file = 's2_strict.fasta', format = 'fasta', nbcol = -1, colsep = '')
```

```r
# relaxed
p1 <- tr1
p1$edge.length <- tr1$edge.length * rlnorm(length(tr1$edge.length), meanlog = log(0.001), sd = 0.1)
s1 <- as.DNAbin(simSeq(p1))
write.dna(s1, file = 's1_relaxed.fasta', format = 'fasta', nbcol = -1, colsep = '')



p2 <- tr2
p2$edge.length <- tr2$edge.length * rlnorm(length(tr2$edge.length), meanlog = log(0.001), sd = 0.1)
s2 <- as.DNAbin(simSeq(p2))
write.dna(s2, file = 's2_relaxed.fasta', format = 'fasta', nbcol = -1, colsep = '')
```
