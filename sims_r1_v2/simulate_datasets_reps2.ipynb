```r
library(ape)
library(TreeSim)
library(phangorn)
library(NELSI)
```

```r
trees <- list()

for(i in 3:10){
    tr <- sim.bd.taxa.age(n = 10, lambda = 1, mu = 0, age = 100, numbsim = 1)[[1]]
    tr$edge.length <- (tr$edge.length / max(allnode.times(tr))) * 100
    
    #strict
    tr_s <- tr
    tr_s$edge.length <- tr$edge.length * 0.001
    ss <- simSeq(tr_s)
    write.dna(as.DNAbin(ss), file= paste0('s', i, '_strict.fasta'), format = 'fasta', nbcol = -1, colsep = '')
    write.dna(as.DNAbin(ss)[, 1:500], file= paste0('s', i, '_strict_subset.fasta'), format = 'fasta', nbcol = -1, colsep = '')
    
    #ucl
    tr_ucl <- tr
    tr_ucl$edge.length <- tr$edge.length * rlnorm(length(tr_ucl$edge.length), meanlog = log(0.001), sd = 0.2)
    sucld <- simSeq(tr_ucl)
    write.dna(as.DNAbin(sucld), file = paste0('s', i, '_relaxed.fasta'), format = 'fasta', nbcol = -1, colsep= '')
    write.dna(as.DNAbin(sucld)[, 1:500], file = paste0('s', i, '_relaxed_subset.fasta'), format = 'fasta', nbcol = -1, colsep = '')
}
```
