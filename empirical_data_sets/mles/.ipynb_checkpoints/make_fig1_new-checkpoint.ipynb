```r
enterovirus <- read.table('enterovirus_ucld_exponential.log', head = T)
wnv <- read.table('wnv_ucld_exponential.log', head = T)
rhdv <- read.table('rhdv_ucld_exponential.log', head = T)
shigella <- read.table('shigella_sonnei_ucld_exponential.log', head = T)
```

```r
colnames(enterovirus)
```

```r
pdf('figure_1_new.pdf', useDingbats = F)
par(mfrow = c(4, 2))
par(mar = c(3, 3, 2, 2))
hist(enterovirus$coefficientOfVariation[100:nrow(wnv)], col = 'grey', main = '', xlab = '', 
     freq = F)
hist(enterovirus$exponential.growthRate[100:nrow(wnv)], col = 'grey', main = '', xlab = '', ylab = '', 
     freq = F)

hist(wnv$coefficientOfVariation[200:nrow(wnv)], col = 'grey', main = '', xlab = '', 
     freq = F)
hist(wnv$exponential.growthRate[100:nrow(wnv)], col = 'grey', main = '', xlab = '', ylab = '', 
     freq = F)

hist(rhdv$coefficientOfVariation[200:nrow(wnv)], col = 'grey', main = '', xlab = '', 
     freq = F, xlim = c(0, 1.2))
hist(rhdv$exponential.growthRate[100:nrow(wnv)], col = 'grey', main = '', xlab = '', ylab = '', 
     freq = F)

hist(shigella$coefficientOfVariation[200:nrow(wnv)], col = 'grey', main = '', xlab = '', 
     freq = F)
hist(shigella$exponential.growthRate[100:nrow(wnv)], col = 'grey', main = '', xlab = '', ylab = '', 
     freq = F)

dev.off()
```
